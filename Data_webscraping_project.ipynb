# Web scraping

From Flipcart data of samsung smartphones scrape data containing smartphone model,rating,customer rating ,smartphone features, price and discount on flipcart.
https://www.flipkart.com/search?q=samsung+smartphone&sid=tyy%2C4io&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_2_11_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_2_11_na_na_na&as-pos=2&as-type=RECENT&suggestionId=samsung+smartphone%7CMobiles&requestId=88479d18-10a7-4944-a2ed-43f81dc7ec03&as-searchtext=samsung+sma&page=1

```python
#importing inbult libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import re
```

```python
#create user defined function all_pages() which takes one passing argument as page number
#and returns value
def all_pages(page_no):
    r=requests.get('''https://www.flipkart.com/search?q=samsung
               +smartphone&sid=tyy%2C4io&as=on&as-show=on&
               otracker=AS_QueryStore_OrganicAutoSuggest_2_11_na_na_na&
               otracker1=AS_QueryStore_OrganicAutoSuggest_2_11_na_na_na&as-pos=2&
               as-type=RECENT&suggestionId=samsung+smartphone%7CMobiles&
               requestId=88479d18-10a7-4944-a2ed-43f81dc7ec03&as-searchtext=samsung+sma&page='''+str(page_no))
    #get() inbuilt method of requests library which takes URL as passing argument
    content=r.content #inbuilt method content to read texts
    soup=BeautifulSoup(content)#inbuilt library BeautifulSoup used for getting data out of HTML,XML.
    L=[]#create empty list
    for d in soup.find_all('div',attrs={'class':'_3pLy-c row'}):#outer loop
        #find_all is inbuilt method to find all div class from html code
        name=d.find('div',attrs={'class':'_4rR01T'})
        #find is inbuilt method to find particular div tag from html code
        star=d.find('div',attrs={'class':'_3LWZlK'})
        rating=d.find('span',attrs={'class':'_2_R_DZ'})
        #find is inbuilt method to find particular span tag from html code
        ul=d.find('ul',attrs={'class':'_1xgFaf'})
        #find is inbuilt method to find particular ul tag from html code
        spec=ul.find_all('li',attrs={'class':'rgWa7D'})
        #find_all is inbuilt method to find all li tag from ul tag 
        ram=spec[0]#1st li tag
        size=spec[1]#2nd li tag
        cam=spec[2]#3rd li tag
        batt=spec[3]#4th li tag
        price=d.find('div',attrs={'class':'_3I9_wc _27UcVY'})
        discount_per=d.find('div',attrs={'class':'_3Ay6Sb'})
        Final_price=d.find('div',attrs={'class':'_30jeq3 _1_WHN1'})
        L1=[]#create empty list
        if name is not None:#condition check for null values
            L1.append(name.text)#inbuilt method append use to add single value to list
            #text inbuilt method to access text value from tag
        else:
            L1.append(np.nan)#nan is inbuilt method of NumPy library to add NaN in place of null values
        if star is not None:
            L1.append(star.text)
        else:
            L1.append(np.nan)
        if rating is not None:#outer if
            rating=rating.text.split() #inbuilt method split() to separate values after blank space
            c=1#initial count
            for value in rating :#inner loop
                if c==1:#inner if
                    L1.append(value)
                c=c+1 #increment
        else:
            L1.append(np.nan)
        if ram is not None:#outer if
            ram=ram.text.split('|')#inbuilt method split() to separate values after |
            c=1 #initial count
            for value in ram:#inner loop
                if c<=2:#inner if
                    L1.append(value)
                c=c+1#increment
        else:
            L1.append(np.nan)
        if size is not None:
            L1.append(size.text)
        else:
            L1.append(np.nan)
        if cam is not None:
            L1.append(cam.text)
        else:
            L1.append(np.nan)
        if batt is not None:
            L1.append(batt.text)
        else:
            L1.append(np.nan)
        if price is not None:
            L1.append(price.text)
        else:
            L1.append(0)
        if discount_per is not None:
            L1.append(discount_per.text)
        else:
            L1.append(0)
        if Final_price is not None:
            L1.append(Final_price.text)
        else:
            L1.append(np.nan)
        L.append(L1)#add list into another list
    return L #return gives output aher function is called
```

```python
data=[]#create empty list
for i in range(1,26):#start=1 stop=26-1=25 step=+1
    data.append(all_pages(i))#calling user defined function
print(data)
```

```python
#create user defined function in single line using lambda function takes one passing argument
function=lambda a : [values for sublist in a for values in sublist]
```

```python
#create dataframe 
#calling user defined lambda function
df=pd.DataFrame(function(data),columns=['Model_name','Ratings','Customers_rated','RAM','Internal_storage','Display_size','Camera','Battery_mAh','Price','Discount','Final_Price'])
#DataFrame is inbult method in pandas library to create dataframe which takes data and column names
df.to_csv('flipcart_samsung_smartphones.csv',index=False,encoding='utf-8')#utf -unicode tramsformation format
#to_csv() is inbuilt method to store data permanently into csv file
```
